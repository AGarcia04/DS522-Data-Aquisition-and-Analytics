Part 1.
1. Create a variable name text and use the text below as the value:

Beautiful is better than ugly.
Explicit is better than implicit.
Simple is better than complex.
Complex is better than complicated.

2. Loop through the text, and remove all the ‘is’, ‘better’, and ‘than’ in the text.
3. Then make sure that all the words are in lower case.
4. Print the final text.


```python
text = """Beautiful is better than ugly.
Explicit is better than implicit.
Simple is better than complex.
Complex is better than complicated."""

for i in range(len(text)):
    text = text.replace('is', '')
    text = text.replace('better', '')
    text = text.replace('than', '')
    text = text.lower()
print(text)
```

Part 2.
1. Create a list of 50 integers and find all the even numbers.
2. Save all the output in a new list.


```python
my_list = list(range(1, 51))
print(my_list)
my_even_list = []
for i in my_list:
    if i % 2 == 0:
        my_even_list.append(i)
print(my_even_list)
```

Part 3.
1. Edit the above program so it will use the values that are provided by users.
2. Have the user input at least 10 values separated by commas or spaces.
3. Then find all the odd numbers in the list.
4. Make sure to save this as a different Python file.


```python
user_list = []
while len(user_list) < 10:
    user_input = input("Please give me at least 10 numbers separated by commas or spaces: ")
    user_list = [int(i) for i in user_input.replace(',', ' ').split()]
    if len(user_list) < 10:
        print("Please enter at least 10 numbers!")

user_odd_list = []
for i in user_list:
    if i % 2 != 0:
        user_odd_list.append(i)

print(user_list)
print(user_odd_list)
```
