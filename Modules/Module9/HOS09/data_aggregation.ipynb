```python
# Create a new column that computes the tip percentage
df["tip_pct"] = df["tip"] / df["total_bill"]
# Create a groupby object using multiple keys, then retrieve the mean of the 'tip_pct'
df.groupby(["day", "smoker"])['tip_pct'].agg('mean')
```

```python
# Create a groupby object using multiple keys, then retrieve the mean, min, max, & std of the 'tip_pct'
df.groupby(["day", "smoker"])['tip_pct'].agg(['mean', 'min', 'max', 'std'])
```

```python
df.groupby(["day", "smoker"])['tip_pct'].agg([('average', 'mean'),('minimum', 'min'),('maximum', 'max'),('std dev', 'std')])
```

```python
# Group by time, then compute the tip's max value and the size's min value
df.groupby('time').agg({'tip': max, 'size': min})
```

```python
df.groupby('time')[['tip', 'size']].apply(lambda x: x[['tip', 'size']].sum(axis=1))
```

```python
new_row = pd.DataFrame({'total_bill': None, 'tip': None,
                        'sex': ['Female'], 'smoker': ['Yes'],
                        'day': ['Friday'], 'time': ['Dinner'],
                        'size': None, 'tip_pct': None})
df = pd.concat([df, new_row])
df
```

```python
def fill_mean(group):
    # Apply fillna with mean only to numeric columns
    numeric_cols = group.select_dtypes(include='number')
    return group.fillna(numeric_cols.mean())

df.groupby('time').apply(fill_mean)
```

```python
df.dropna(inplace=True)
df.groupby('time')[['total_bill', 'tip', 'size', 'tip_pct']].transform(lambda x: x + x.std())
```

```python
df.groupby('time').filter(lambda x: x['size'].std() > 1)
```

```python
pd.pivot_table(df, index='time', columns='sex', values='tip', aggfunc=sum, fill_value=0)
```
