```python
import numpy as np
import pandas as pd
import seaborn as sb
import matplotlib.pyplot as plt
```

```python
x = np.linspace(-np.pi, np.pi, 256, endpoint=True)
c, s = np.cos(x), np.sin(x)
```

```python
plt.plot(x, c, '-c')   # - means solid line and c means cyan color
plt.plot(x, s, '--m')  # -- means dashed line and m means magenta color
```

```python
plt.plot(x, c, color="cyan", linestyle="-")
plt.plot(x, s, color="magenta", linestyle="--")
```

```python
# Initialize a figure
fig = plt.figure()
# Add axes to the figure
ax = fig.add_axes([0, 0, 1, 1])

y = [1, 4, 9, 16, 25, 36, 49, 64]
x1 = [1, 16, 30, 42, 55, 68, 77, 88]  # AMZ
x2 = [1, 6, 12, 18, 28, 40, 52, 65]   # MSFT

l1 = ax.plot(x1, y, 'bs-')   # solid line with blue color and square marker
l2 = ax.plot(x2, y, 'ro--')  # dashed line with red color and circle marker
ax.legend(labels=('AMZ', 'MSFT'), loc='lower right')
ax.set_title('Stock Market')
ax.set_xlabel('Time')
ax.set_ylabel('USD')
plt.show()
```

```python
x = np.arange(0, np.pi*2, 0.05)
y = np.sin(x)
z = np.cos(x)

# Create a figure
fig2 = plt.figure()

# Set up axes
axes1 = fig2.add_axes([0.1, 0.1, 0.8, 0.8])   # main axes (blue)
axes2 = fig2.add_axes([0.55, 0.55, 0.3, 0.3])  # inserted axes (red)

# Plotting
axes1.plot(x, y, 'b')
axes2.plot(x, z, 'r')
axes1.set_title('sine')
axes2.set_title('cosine')
plt.show()
```

```python
def f(x, y):
    return (1 - x / 2 + x ** 5 + y ** 3) * np.exp(-x ** 2 - y ** 2)

n = 256
x = np.linspace(-3, 3, n)
y = np.linspace(-3, 3, n)
X, Y = np.meshgrid(x, y)

plt.contourf(X, Y, f(X, Y), 8, alpha=.75, cmap='jet')
C = plt.contour(X, Y, f(X, Y), 8, colors='black')
plt.show()
```

```python
n = 12
x = np.arange(n)
y1 = (1 - x / float(n)) * np.random.uniform(0.5, 1.0, n)
y2 = (1 - x / float(n)) * np.random.uniform(0.5, 1.0, n)

plt.bar(x, +y1, facecolor='#9999ff', edgecolor=['green', 'red'])
plt.bar(x, -y2, facecolor='#ff9999', edgecolor=['white'])

for x, y in zip(x, y1):
    # Text properties and layout
    plt.text(x + 0.4, y + 0.05, f"{y:.2f}", ha='center', va='center')

plt.ylim(-1.25, +1.25)
plt.show()
```

```python
from mpl_toolkits.mplot3d import Axes3D

fig = plt.figure()
ax = Axes3D(fig)
fig.add_axes(ax)

x = np.arange(-4, 4, 0.25)
y = np.arange(-4, 4, 0.25)

# Meshgrid returns coordinate matrices from coordinate vectors
x, y = np.meshgrid(x, y)
r = np.sqrt(x ** 2 + y ** 2)
z = np.sin(r)

ax.plot_surface(x, y, z, rstride=1, cstride=1, cmap='winter')

plt.show()
```

```python
# load tips dataset from seaborn
df = sb.load_dataset('tips')
sb.relplot(x='total_bill', y='tip', data=df)
# Scatter is default. Add the parameter kind='line' to change to a line plot
```

```python
sb.relplot(x='total_bill', y='tip', hue=df.day.tolist(), data=df)
```

```python
# Load flights dataset and convert to long-form
flights_long = sb.load_dataset('flights')
flights = flights_long.pivot(index='month', columns='year', values='passengers')

# Draw a heatmap with the numeric values in each cell
f, ax = plt.subplots(figsize=(12, 9))
sb.heatmap(flights, annot=True, fmt='d', linewidth=.5, ax=ax)
```
