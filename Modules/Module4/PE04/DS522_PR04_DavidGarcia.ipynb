```python
import pandas as pd

# Load the CSV into a DataFrame
sales_df = pd.read_csv('/Users/alexgarcia/Desktop/CityUniversity/MSDS/DS522 Data Aquisition and Analytics/Modules/Module4/PE04/sales.csv')
sales_df
```

```python
# Summary information
print("Summary Information:")
sales_df.info()
```

```python
# Summary statistics for numerical columns
print("Summary Statistics:")
sales_df.describe()
```

```python
# Add Total Sales column (Quantity * Price)
sales_df['Total Sales'] = sales_df['Quantity'] * sales_df['Price']

# Convert Date column to datetime
sales_df['Date'] = pd.to_datetime(sales_df['Date'])

print("DataFrame with Total Sales column and datetime Date:")
sales_df
```

```python
# Filter sales that occurred in July
july_sales = sales_df[sales_df['Date'].dt.month == 7]
print("Sales in July:")
print(july_sales)
```

```python
# Filter sales where Quantity > 10
high_qty = sales_df[sales_df['Quantity'] > 10]
print("Sales where Quantity > 10:")
print(high_qty)
```

```python
# Unique store names and total sales per store
print("Unique Stores:", sales_df['Store'].unique())
print("\nTotal Sales per Store:")
print(sales_df.groupby('Store')['Total Sales'].sum())
```

```python
# Unique products and average price per product
print("Unique Products:", sales_df['Product'].unique())
print("\nAverage Price per Product:")
print(sales_df.groupby('Product')['Price'].mean())
```

```python
# Sort by Total Sales descending
sorted_df = sales_df.sort_values('Total Sales', ascending=False)
print("DataFrame sorted by Total Sales (descending):")
print(sorted_df)
```

```python
# Rank stores based on Total Sales
store_totals = sales_df.groupby('Store')['Total Sales'].sum()
store_totals_df = store_totals.reset_index()
store_totals_df['Rank'] = store_totals_df['Total Sales'].rank(ascending=False).astype(int)
store_totals_df = store_totals_df.sort_values('Rank')
print("Store Rankings by Total Sales:")
print(store_totals_df)
```

```python
# Custom function: 10% discount if Quantity > 5
def discounted_price(row):
    if row['Quantity'] > 5:
        return row['Price'] * 0.90
    return row['Price']

# Apply the function to create Discounted Price column
sales_df['Discounted Price'] = sales_df.apply(discounted_price, axis=1)

print("DataFrame with Discounted Price column:")
sales_df
```
