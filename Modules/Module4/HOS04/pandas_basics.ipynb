```python
# Creating a Pandas DataFrame from a dictionary
data_dict = {
    'Name': ['Alice', 'Bob', 'Charlie', 'David'],
    'Age': [25, 30, 35, 40],
    'City': ['New York', 'Los Angeles', 'Chicago', 'Houston']
}

df = pd.DataFrame(data_dict)

print("\nPandas DataFrame:")
print(df)
```

```python
# Creating a DataFrame from lists
names  = ['Alice', 'Bob', 'Charlie', 'David']
ages   = [25, 30, 35, 40]
cities = ['New York', 'Los Angeles', 'Chicago', 'Houston']

df_from_list = pd.DataFrame({'Name': names, 'Age': ages, 'City': cities})

print("\nDataFrame from lists:")
print(df_from_list)
```

```python
df_from_list.set_index("Name", inplace=True)
df_from_list
```

```python
# Create a copy of the original df_from_list
df = df_from_list.copy()

# Accessing a single column
ages = df['Age']
print("\nAccessing a single column:")
print(ages)

# Access data of index 'Alice'
print("\nAccessing a single row:")
df.loc['Alice']
```

```python
# Change back to the default index and store the another df variable
df = df.reset_index()
print("\nAccessing multiple columns:")
# Accessing multiple columns
df[['Name', 'City']]
```

```python
# Creating a DataFrame
companies = {
    'Value': [220, 160, 140, 117, 94],
    'Change': [17.5, 11.9, -8.5, -2.1, 3.5],
    'Sector': ['Retail', 'Tech', 'Tech', 'Tech', 'Tech']
}

df = pd.DataFrame(companies, index=['Amazon', 'Google', 'Apple', 'Microsoft', 'Samsung'])
df
```

```python
# Adding a scalar to all elements
print("\nDataFrame after adding 10 to all elements of the Value column:")
print(df['Value'].add(10))

# Subtracting a scalar from all elements
print("\nDataFrame after subtracting 5 from all elements of the Value column:")
print(df['Value'].sub(5))

# Multiplying all elements by a scalar
print("\nDataFrame after multiplying all elements of the Value column by 2:")
print(df['Value'].mul(2))

# divide and mod values in 'Value' by 2
div, rem = divmod(df['Value'], 2)
print("\nDataFrame after dividing all elements by 5:")
div, rem
# div is the floor of each values divided by 2
# rem is the remainder of each value divided by 2
```

```python
# Sum of all elements in the DataFrame
total_sum = df.sum()
print("\nSum of all elements in each column:")
print(total_sum)
```

```python
# Sum along the rows, pass numeric_only=True to avoid errors
row_sum = df.sum(axis=1, numeric_only=True)
print("\nSum of elements across each row:")
print(row_sum)
```

```python
# Mean of each column
mean_values = df.mean(numeric_only=True)
print("\nMean of each column:")
print(mean_values)

# Median of each column
median_values = df.median(numeric_only=True)
print("\nMedian of each column:")
print(median_values)

# Minimum values of each column
min_values = df.min(numeric_only=True)
print("\nMinimum value in each column:")
print(min_values)

# Maximum values of each column
max_values = df.max(numeric_only=True)
print("\nMaximum value in each column:")
print(max_values)
```

```python
# Standard deviation of each column
std_values = df.std(numeric_only=True)
print("\nStandard deviation of each column:")
print(std_values)

# Variance of each column
variance_values = df.var(numeric_only=True)
print("\nVariance of each column:")
print(variance_values)

# The 50% quantile
quant_values = df.quantile(0.5, numeric_only=True)
print("\nThe 50% Quantile of each column:")
print(quant_values)
```

```python
# Cumulative sum of each column
cumsum_values = df[['Value', 'Change']].cumsum()
print("\nCumulative sum of each column:")
print(cumsum_values)

# Cumulative product of each column
cumprod_values = df[['Value', 'Change']].cumprod()
print("\nCumulative product of each column:")
print(cumprod_values)
```

```python
# Correlation between columns
correlation_matrix = df.corr(numeric_only=True)
print("\nCorrelation matrix:")
print(correlation_matrix)

# Covariance between columns
covariance_matrix = df.cov(numeric_only=True)
print("\nCovariance matrix:")
print(covariance_matrix)
```

```python
# Define a custom function
def double_values(x):
    return x * 2

# Apply the custom function to the DataFrame
df_custom = df.apply(double_values)
print("\nDataFrame after applying custom function (doubling values):")
print(df_custom)
```
