```python
import sqlite3

# Create a connection to the SQLite database (or create the database if it doesn't exist)
conn = sqlite3.connect('sales_data.db')

# Create a cursor object to interact with the database
cursor = conn.cursor()

# Create a sales table
cursor.execute('''
    CREATE TABLE IF NOT EXISTS sales_table (
        id INTEGER PRIMARY KEY,
        Date TEXT,
        Store TEXT,
        Product TEXT,
        Quantity INTEGER,
        Price REAL
    )
''')

# Commit the changes
conn.commit()

# Close the connection
conn.close()
```

```python
import sqlite3

# Sample data to insert into the table
sales_data = [
    ('2023-07-01', 'Store A', 'Product 1', 5, 10.0),
    ('2023-07-02', 'Store B', 'Product 2', 3, 15.0),
    ('2023-07-03', 'Store A', 'Product 1', 2, 10.0),
    ('2023-07-04', 'Store C', 'Product 3', 8, 20.0),
    ('2023-07-05', 'Store B', 'Product 2', 1, 15.0)
]

# Reconnect to the SQLite database
conn = sqlite3.connect('sales_data.db')

# Create a cursor object
cursor = conn.cursor()

# Insert data into the sales_table
cursor.executemany('''
    INSERT INTO sales_table (Date, Store, Product, Quantity, Price)
    VALUES (?, ?, ?, ?, ?)
''', sales_data)

# Commit the changes
conn.commit()

# Close the connection
conn.close()
```

```python
import pandas as pd
import sqlite3

# Sample DataFrame to write to the database
data = {
    'Date': ['2023-07-06', '2023-07-07'],
    'Store': ['Store D', 'Store E'],
    'Product': ['Product 4', 'Product 5'],
    'Quantity': [4, 7],
    'Price': [25.0, 30.0]
}
df_new = pd.DataFrame(data)

# Reconnect to the SQLite database
conn = sqlite3.connect('sales_data.db')

# Write the DataFrame to the sales_table
df_new.to_sql('sales_table', conn, if_exists='append', index=False)

# Commit the changes (not necessary for `to_sql`, but is a good practice)
conn.commit()

# Close the connection
conn.close()

print("New data has been added to the database.")
```

```python
import pandas as pd
import sqlite3

# Reconnect to the SQLite database
conn = sqlite3.connect('sales_data.db')

# Load data from the sales_table into a Pandas DataFrame
df = pd.read_sql_query('SELECT * FROM sales_table', conn)

# Display the first few rows
print(df.head())

# Close the connection
conn.close()
```
