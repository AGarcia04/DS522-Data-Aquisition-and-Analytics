```python
import pandas as pd
import numpy as np
```

## Data Preparation and Cleaning
- Load the 2 datasets: 'solid_waste_king' and 'census'
- Remove irrelevant rows and columns. Only focus on 2015-2019
- Handle missing data
- Merge the 2 datasets using County Code


```python
# Load census data
df_pop = pd.read_excel('census.xlsx', sheet_name='Data', header=0, skiprows=[1]) # header as column names, skip the second sup header
# Drop irrelevant columns, only focus on 2015-2018
columns_to_drop = np.r_[1:8, 12] # Creates a flat index array: [1,2,3,4,5,6,7,12]
df_pop = df_pop.drop(df_pop.columns[columns_to_drop], axis=1)
# dropping row containing Oregon data
df_pop = df_pop.drop(df_pop.index[0])
df_pop.head()
```

```python
# Rename 'Unnamed: 0' to 'County' for clarity
df_pop = df_pop.rename(columns={'Unnamed: 0': 'County'})
# Remove "County, Washington" from the county names for merging compatability
df_pop['County'] = df_pop['County'].str.replace('County, Washington', '', case=False)
# Removing leading or trailing space
df_pop['County'] = df_pop['County'].str.strip() 
df_pop.head()
```

```python
# Rename columns for clarity
df_pop = df_pop.rename(columns={
    '7/1/2015 population estimate': 'Population_2015',
    '7/1/2016 population estimate': 'Population_2016',
    '7/1/2017 population estimate': 'Population_2017',
    '7/1/2018 population estimate': 'Population_2018'
})
df_pop
```

```python
# Check for missing values
df_pop.isnull().sum()
```

```python
df_pop.info()
```

```python
# Load and inspect waste data for 2015
df_waste_2015 = pd.read_excel('solid_waste.xlsx', sheet_name='2015', header=1, skiprows=[0])
# rename first column for clarity
df_waste_2015.rename(columns={df_waste_2015.columns[0]: 'County', df_waste_2015.columns[-1]: 'Grand_Total'}, inplace=True)
df_waste_2015
```

This dataset has two-level structure — county rows (like "Adams", "Asotin") with facility sub-rows underneath. The Grand Total tons of waste are one the same rows where the county names are, we would need to specify the county names, and show the total waste of each county

```python
# Get a list of county names from the df_pop DF
counties = df_pop['County'].tolist() 
counties
```

```python
# Match county names from df_waste_2015 to df_pop, return a new DF
data2015 = df_waste_2015[df_waste_2015['County'].isin(counties)]
data2015.head(10)
```

For the purpose of this analysis, we can skip removing missing values as they provide helpful insights on the specific types of waste which counties produce. Further analysis will show the trends of types of waste each county produces

```python
# Load and inspect waste data for 2016
df_waste_2016 = pd.read_excel('solid_waste.xlsx', sheet_name='2016', header=1, skiprows=[0])
# rename first column for clarity
df_waste_2016.rename(columns={df_waste_2016.columns[0]: 'County', df_waste_2016.columns[-1]: 'Grand_Total'}, inplace=True)
# Match county names from df_waste to df_pop, return a new DF
data2016 = df_waste_2016[df_waste_2016['County'].isin(counties)]
data2016.head(10)
```

```python
# Load and inspect waste data for 2017
df_waste_2017 = pd.read_excel('solid_waste.xlsx', sheet_name='2017', header=1)
# rename first column for clarity
df_waste_2017.rename(columns={df_waste_2017.columns[0]: 'County', df_waste_2017.columns[-1]: 'Grand_Total'}, inplace=True)
# Match county names from df_waste to df_pop, return a new DF
data2017 = df_waste_2017[df_waste_2017['County'].isin(counties)]
data2017.head(10)
```

```python
# Load and inspect waste data for 2018
df_waste_2018 = pd.read_excel('solid_waste.xlsx', sheet_name='2018', header=1, skiprows=[0])
# rename first column for clarity
df_waste_2018.rename(columns={df_waste_2018.columns[0]: 'County', df_waste_2018.columns[-1]: 'Grand_Total'}, inplace=True)
# Match county names from df_waste to df_pop, return a new DF
data2018 = df_waste_2018[df_waste_2018['County'].isin(counties)]
data2018.head(10)
```

## Merge all the datasets
We want to merge the waste data from 2015-2018 to the population data for the same years with the key is 'County'.

```python
# Rename Grand_Total column in each year's df before merging
data2015_m = data2015[['County', 'Grand_Total']].rename(columns={'Grand_Total': 'Total_2015'})
data2016_m = data2016[['County', 'Grand_Total']].rename(columns={'Grand_Total': 'Total_2016'})
data2017_m = data2017[['County', 'Grand_Total']].rename(columns={'Grand_Total': 'Total_2017'})
data2018_m = data2018[['County', 'Grand_Total']].rename(columns={'Grand_Total': 'Total_2018'})

# Merge all four waste datasets together on County
df_merged = data2015_m \
    .merge(data2016_m, on='County', how='outer') \
    .merge(data2017_m, on='County', how='outer') \
    .merge(data2018_m, on='County', how='outer')

# Merge with population data
df_final = df_merged.merge(df_pop, on='County', how='left')

# Reorder columns
df_final = df_final[['County','Total_2015', 'Population_2015',
                    'Total_2016', 'Population_2016',
                    'Total_2017', 'Population_2017',
                    'Total_2018', 'Population_2018']]

print(df_final.head())
print(df_final.shape)
```
